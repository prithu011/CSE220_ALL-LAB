```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row,col=floor.shape

  for i in range(col):
      if i%2==0:
         for j in range(row):
             if j%2==0:
                print(floor[j][i], end=" ")
         print()

      elif i%2!=0:
         for k in range(row-1,-1,-1):
             if k%2!=0:
                print(floor[k][i], end=" ")
         print()



floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    row, col = district.shape
    array= np.zeros((row+2*depth,col+2*depth),dtype=int)
    for i in range(row+2*depth):
        for j in range(col+2*depth):
            if((i >= depth) and (i < depth+row) and (j >= depth) and (j < depth+col)):
                array[i][j] = district[i-depth][j-depth]
            else:
                array[i][j] = 8
    return array


depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
    n = clubroom.shape[0]
    len_strength_diff = (n**2 - n) //2 #Bonus problem
    diff = np.zeros(len_strength_diff,dtype=int)
    count = 0
    for i in range(n):
        for j in range(i,n):
            if(i == j):
                continue
            diff[count] = clubroom[i][j] - clubroom[j][i]
            count += 1
    return diff


clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
    arr = np.zeros((n,n),dtype=int)
    arr[0][0] = 1
    for i in range(1,n):
        sum = 0
        for j in range(0,i+1):
            sum += arr[i-1][j]
            if((j == 0) or (j == i)):
                arr[i][j] = i + 1
            else:
                arr[i][j] = sum
    return arr


def print_without_none(matrix):
    n = matrix.shape[0]
    for i in range(n):
        for j in range(n):
            if(matrix[i][j] == 0):
                break
            print(matrix[i][j],end=" ")
        print()



n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def check(a,b):
    if((0 <= a <= 7) and (0 <= b <= 7)):
        return True
    else:
        return False

def func(a,b,cmd):
    x = a
    y = b
    if(cmd == 1):
        x = x - 3
        y = y - 2
    elif(cmd == 2):
        x = x - 1
        y = y - 2
    elif(cmd == 3):
        x = x - 2
        y = y - 3
    elif(cmd == 4):
        x = x - 2
        y = y - 1
    elif(cmd == 5):
        x = x + 1
        y = y - 2
    elif(cmd == 6):
        x = x + 3
        y = y - 3
    elif(cmd == 7):
        x = x + 2
        y = y - 3
    elif(cmd == 8):
        x = x + 2
        y = y - 1
    elif(cmd == 9):
        x = x - 3
        y = y + 2
    elif(cmd == 10):
        x = x - 1
        y = y + 2
    elif(cmd == 11):
        x = x - 2
        y = y + 1
    else:
        x = x - 2
        y = y + 3
    return x,y

def moving_around(cmds):
    grid = np.full((7, 7), '.')
    grid[3][3] = "-"
    x,y = 3,3
    for i in range(len(cmds)):
        a,b = func(x,y,cmds[i])
        if(check(a,b)):
            x,y = a,b
            grid[y][x] = "*"
    grid[y][x] = "/"
    return grid

cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
